```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ejercicio 1

def euler_explicito(f, y_0, t_0, t_f, h):
    t = t_0
    y = y_0
    vector_t = []
    vector_y = []

    while t < t_f:
        vector_t.append(t)
        vector_y.append(y)
        y = y + h * f(t, y)
        t = t + h

    return (vector_t, vector_y)
```

```python
# Ejercicio 2
f = lambda t, y: (y - 5) * (np.cos(t)**2 - 0.5)
T = 10

for k in range(11):
    plt.plot(*euler_explicito(f, k, 0, T, 0.01))
```
